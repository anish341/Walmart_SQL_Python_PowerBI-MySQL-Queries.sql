# Handling Categorical Data


```python
## One-hot encoding of categorical variables.

df_encoded = pd.get_dummies(df, columns=["date"])
print(df_encoded)

```

```python
# # Converts categorical labels to numeric
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
df["category"] =le.fit_transform(df["category"])
print(df)
```

```python
df['category_code'], uniques = pd.factorize(df['category'])
print(df)
```

# Scaling and Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Applying StandardScaler to the 'category' column
df["category"] = scaler.fit_transform(df[["category"]])

# Displaying the updated DataFrame
print(df)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df['category'] = scaler.fit_transform(df[['category']])
print(df)
```

```python
from sklearn.preprocessing import normalize
df['category'] = normalize(df[['category']])
print()
```

```python
df['unit_price'] = df['City'].apply(lambda x: x * 2)
print(df)
```

```python
df["new_col"] =df["unit_price"].map({1:"a",2:"b"})
print(df)
```


# Feature Selection

```python
# #It is use of unwanted columns
df = df.drop(["new_col"],axis =1)
```
